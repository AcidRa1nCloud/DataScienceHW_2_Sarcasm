```python
# Tools for text Normalization
stop_words = stop_words = set(stopwords.words('english')).union({'also', 'would', 'much', 'many'})
stop_words = stop_words.difference(negations)
stop_words = stop_words.union({'.', ',', ';', ':', '\'', '?', '*', '(', ')', '#', '$', '--'})

nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
stemmer = PorterStemmer()
```

```python
def text_processor(
    text: str,
    type: str='lemma',
    stop_words: set=stop_words,
    contractions: dict=contractions
    ) -> str:
    """
    Normalizes the text to a more suitable for work way

    Args:
        text: text that we want to preprocess
        type: select a type of converting words lemma/stem
        stop_words: a set of stop_words
        contractions: a dictionary of contractions

    Returns:
        A string of tokens
    """
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'m", " am", text)
    text = re.sub(r"\d+\w*", '', text)

    text = word_tokenize(text.lower())
    text = list(map(lambda x: contractions[x] if x in contractions else x, text))
    text = list(filter(lambda y: y not in stop_words, text))

    match type:
        case 'lemma':
            text = ' '.join(text)
            doc = nlp(text)
            text = ' '.join([token.lemma_ for token in doc if len(token.lemma_) > 1])
        case 'stem':
            text = list(map(lambda z: stemmer.stem(z), text))
            text = ' '.join(text)
        case _:
            raise Exception('Type Attribute has to be: stem or lemma.')

    return text
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report
```

```python
def train_val_split(
    X: np.array,
    y: np.array,
    train_split: int=70
    ):
    """
    Acts like train_test_split() from sklearn + validation split

    Args:
        X: Feature matrix
        y: Target vector
        train_split: percent of train split

    Returns:
        Like sklearn function returns slices into train and test data
    """
    val_split = (100 - train_split) // 2
    t = len(y)
    p_s = t*train_split//100
    p_f = (t*val_split//100)+p_s
    X_train, y_train, X_val, y_val, X_test, y_test = X[:p_s], y[:p_s], X[p_s:p_f], y[p_s:p_f], X[p_f:], y[p_f:]
    return X_train, y_train, X_val, y_val, X_test, y_test

def print_metrics(
    ground_truth: np.array,
    predictions: np.array
    ):
    """
    Prints out multiple metrics of the model

    Args:
        ground_truth: our target feature
        predictions: Our predictions
    """
    print("Classification report:\n", classification_report(ground_truth, predictions),'\n')
    print("AUC:\n", roc_auc_score(ground_truth, predictions), '\n')

def model_algo_selection(
    model: str,
    X_train: np.array,
    X_val: np.array,
    y_train: np.array,
    y_val: np.array,
    algo:str='cv',
    ngram: tuple=(1, 1)
    ):
    """
    Fast way of trying out different combos of algorithms

    Args:
        model: sklearn model object
        algo: nlp algorithm such as countVector as cv or tf-idf
        ngram: a tuple with the size of n-gram
        X_train: training data
        X_val: validation data
        y_train: training target
        y_val: validation target
    """
    model = model()

    match algo:
        case 'cv':
            vect = CountVectorizer(ngram_range=ngram).fit(X_train)
            X_train = vect.transform(X_train)
            X_val = vect.transform(X_val)
        case 'tf-idf':
            vect = TfidfVectorizer(ngram_range=ngram).fit(X_train)
            X_train = vect.transform(X_train)
            X_val = vect.transform(X_val)
        case 'embedded':
            pass
    model.fit(X_train, y_train)
    predictions = model.predict(X_val)
    print_metrics(y_val, predictions)
```

```python
class WordEmbedding:
    """
    Simplifies our work with multiple models
    """
    def __init__(self, verbose=0):
        self.verbose = verbose
        self.model = {}


    def convert(self, source, input_file_path, output_file_path):
        """
        Converts glove model from text document to binary
        """
        match source:
            case 'glove':
                input_file = datapath(input_file_path)
                output_file = get_tmpfile(output_file_path)
                glove2word2vec(input_file, output_file)
            case 'word2vec':
                pass
            case 'fasttext':
                pass
            case 'from_scratch':
                pass
            case _:
                raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')


    def load(self, source, filepath):
        """
        Loads a model from a given file path
        """
        print(datetime.datetime.now(), 'start: loading', source,'\n')
        match source:
            case ('glove' | 'fasttext'):
                self.model[source] = gensim.models.KeyedVectors.load_word2vec_format(filepath)
            case ('word2vec' | 'from_scratch'):
                self.model[source] = gensim.models.KeyedVectors.load_word2vec_format(filepath, binary=True)
            case _:
                raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        print(datetime.datetime.now(), 'end: loading', source,'\n')
        return self


    def get_model(self, source):
        """
        Returns a model object from a dictionary of models
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        return self.model[source]


    def get_words(self, source, size=None):
        """
        Returns a list of words known to a model
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if size is None:
            return [w for w in self.get_model(source=source).key_to_index]
        else:
            results = []
            for i, word in enumerate(self.get_model(source=source).key_to_index):
                if i >= size:
                    break
                results.append(word)
            return results
        return Exception('Unexpected flow')


    def get_dimension(self, source):
        """
        Returns the number of dimensions
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')

        return self.get_model(source=source).vectors[0].shape[0]


    def get_vectors(self, source, words=None):
        """
        Returns vectors representation of multiple words
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if words is None:
            words = self.get_words(source=source)
        embedding = np.empty((len(words), self.get_dimension(source=source)), dtype=np.float32)
        for i, word in enumerate(words):
            embedding[i] = self.get_vector(source=source, word=word)
        return embedding


    def get_vector(self, source, word):
        """
        Returns vector representation of a single word
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if source not in self.model:
            raise ValueError('Did not load %s model yet' % source)
        try:
            return self.model[source][word]
        except KeyError as e:
            dims = self.model[source][0].shape
            vect = np.empty(dims)
            vect[:] = np.nan
            return vect


    def get_synonym(self, source, word, topn=5):
        """
        Returns similar words of a given words based on the vector distance between them
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if source not in self.model:
            raise ValueError('Did not load %s model yet' % source)
        try:
            return self.model[source].most_similar(positive=word, topn=topn)
        except KeyError as e:
            print(e,'\n')


    def get_distance_between_two_words(self, source, word1, word2):
        """
        Calculates the distance between two vectors each of which represents a word
        """
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if source not in self.model:
            raise ValueError('Did not load %s model yet' % source)
        try:
            return self.model[source].similarity(word1, word2)
        except KeyError as e:
            raise
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format(word2vec_path, binary=True)
```

```python
keys = ['python', 'programming', 'inteligence', 'agent', 'danger', 'animal', 'fire', 'water', 'war', 'money', 'toxic', 'drive', 'alphabet', 'pepsi']

word_cluster = []
embedding_cluster = []

for k in keys:
    similar = list(dict(model.most_similar(k, topn=10)).keys())
    word_cluster.append(similar)
    embeddings = model[similar]
    embedding_cluster.append(embeddings)
```

```python
word_cluster = np.array(word_cluster)
embedding_cluster = np.array(embedding_cluster)
```

**Arguments:**

`perplexity` - perplexity is related to the number of nearest neighbours that is used in other manifold learning algorithms

`n_components` - Dimensions of the embedded space

`init` - type of initialization mostly used 'PCA'

`n_iter` - Maximum number of iterations for the optimization (min=250)

```python
n, m, k = embedding_cluster.shape

tsne_model_en_2d = TSNE(
    perplexity=15,
    n_components=2,
    init='pca',
    n_iter=3500
)

embeddings_en_2d = np.array(tsne_model_en_2d.fit_transform(embedding_cluster.reshape(n * m, k))).reshape(n, m, 2)
```

```python
def plot_tsne(
    title: str, 
    alpha: float=1, 
    word_clusters: np.array=word_cluster, 
    embedding_clusters: np.array=embeddings_en_2d, 
    labels: list=keys, 
    filename: str=None
    ):
    """
    Plots the resul of TSNE model
    
    Args:
        title: title of the plot
        labels: target words
        embedding_cluster: list of vector representations of words
        word_cluster: list of similar words
        alpha: transperency
        filename: name of file for saving the image of the plot
    """
    plt.figure(figsize=(20, 12))
    colors = cm.rainbow(np.linspace(0, 1, len(labels)))
    for label, embedding, word, color in zip(labels, embedding_clusters, word_clusters, colors):
        x = embedding[:, 0]
        y = embedding[:, 1]
        plt.scatter(x, y, color=color, alpha=alpha, label=label)
        for i, word in enumerate(word):
            plt.annotate(word, alpha=0.5, xy=(x[i], y[i]), xytext=(5, 2),
                        textcoords='offset points', ha='right', va='bottom', size=8)
    plt.legend(loc=4)
    plt.title(title)
    plt.grid(True)
    if filename:
        plt.savefig(filename, format='png', dpi=150, bbox_inches='tight')
    plt.show()
```

```python
def tokens_vector(
    tokens: list,
    source: str,
    avg: str
    ):
    """
    Returns a vector representation of a list with tokens based on some arithmetic function
    
    Args:
        tokens: list of tokens
        source: name of model we are using
        avg: mathmatical operation which will be used to represent multiple vectors as one
    """
    vects = word_embedding.get_vectors(source=source, words=tokens)
    match avg:
        case 'mean':
            return np.nanmean(vects, axis=0)
        case 'median':
            return np.nanmedian(vects, axis=0)
        case 'sum':
            return np.nansum(vects, axis=0)
        case _:
            raise ValueError('Select correct avereging method: sum, median, mean')
```
