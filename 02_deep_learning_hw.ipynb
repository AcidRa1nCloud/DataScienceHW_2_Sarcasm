```python
# Tools for text Normalization
stop_words = stop_words = set(stopwords.words('english')).union({'also', 'would', 'much', 'many'})
stop_words = stop_words.difference(negations)
stop_words = stop_words.union({'.', ',', ';', ':', '\'', '?', '*', '(', ')', '#', '$', '--'})

nlp = spacy.load("en_core_web_sm", disable=['parser', 'ner'])
stemmer = PorterStemmer()
```

```python
def text_processor(
    text: str,
    type: str='lemma',
    stop_words: set=stop_words,
    contractions: dict=contractions
    ) -> str:
    """
    Normalizes the text to a more suitable for work way

    Args:
        text: text that we want to preprocess
        type: select a type of converting words lemma/stem
        stop_words: a set of stop_words
        contractions: a dictionary of contractions

    Returns:
        A string of tokens
    """
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\'m", " am", text)
    text = re.sub(r"\d+\w*", '', text)

    text = word_tokenize(text.lower())
    text = list(map(lambda x: contractions[x] if x in contractions else x, text))
    text = list(filter(lambda y: y not in stop_words, text))

    match type:
        case 'lemma':
            text = ' '.join(text)
            doc = nlp(text)
            text = ' '.join([token.lemma_ for token in doc if len(token.lemma_) > 1])
        case 'stem':
            text = list(map(lambda z: stemmer.stem(z), text))
            text = ' '.join(text)
        case _:
            raise Exception('Type Attribute has to be: stem or lemma.')

    return text
```

```python
df['normalized_text'] = df.headline.apply(text_processor)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, classification_report
```

```python
def train_val_split(
    X: np.array,
    y: np.array,
    train_split: int=70
    ):
    """
    Acts like train_test_split() from sklearn + validation split

    Args:
        X: Feature matrix
        y: Target vector
        train_split: percent of train split

    Returns:
        Like sklearn function returns slices into train and test data
    """
    val_split = (100 - train_split) // 2
    t = len(y)
    p_s = t*train_split//100
    p_f = (t*val_split//100)+p_s
    X_train, y_train, X_val, y_val, X_test, y_test = X[:p_s], y[:p_s], X[p_s:p_f], y[p_s:p_f], X[p_f:], y[p_f:]
    return X_train, y_train, X_val, y_val, X_test, y_test

def print_metrics(
    ground_truth: np.array,
    predictions: np.array
    ):
    """
    Prints out multiple metrics of the model

    Args:
        ground_truth: our target feature
        predictions: Our predictions
    """
    print("Classification report:\n", classification_report(ground_truth, predictions),'\n')
    print("AUC:\n", roc_auc_score(ground_truth, predictions), '\n')

def model_algo_selection(
    model: str,
    X_train: np.array,
    X_val: np.array,
    y_train: np.array,
    y_val: np.array,
    algo:str='cv',
    ngram: tuple=(1, 1)
    ):
    """
    Fast way of trying out different combos of algorithms

    Args:
        model: sklearn model object
        algo: nlp algorithm such as countVector as cv or tf-idf
        ngram: a tuple with the size of n-gram
        X_train: training data
        X_val: validation data
        y_train: training target
        y_val: validation target
    """
    model = model()

    match algo:
        case 'cv':
            vect = CountVectorizer(ngram_range=ngram).fit(X_train)
        case 'tf-idf':
            vect = TfidfVectorizer(ngram_range=ngram).fit(X_train)
    model.fit(vect.transform(X_train), y_train)
    predictions = model.predict(vect.transform(X_val))
    print_metrics(y_val, predictions)
```

```python
class WordEmbedding:


    def __init__(self, verbose=0):
        self.verbose = verbose=0
        self.model = {}


    def convert(self, source, input_file_path, output_file_path):
        match source:
            case 'glove':
                input_file = datapath(input_file_path)
                output_file = get_tmpfile(output_file_path)
                glove2word2vec(input_file, output_file)
            case 'word2vec':
                pass
            case 'fasttext':
                pass
            case 'from_scratch':
                pass
            case _:
                raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')


    def load(self, source, file_path):
        print(datetime.datetime.now(), 'start: loading', source)
        match source:
            case 'glove' | 'fasttext':
                self.model[source] = gensim.models.KeyedVectors.load_word2vec_format(filepath)
            case 'word2vec' | 'from_scratch':
                self.model[source] = gensim.models.KeyedVectors.load_word2vec_format(filepath, binary=True)
            case _:
                raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        print(datetime.datetime.now(), 'end: loading', source)
        return self


    def get_model(self, source):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        return self.model[source]


    def get_words(self, source, size=None):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if size is None:
            return [w for w in self.get_model(source=source).key_to_index]
        else:
            results = []
            for i, word in enumerate(self.get_model(source=source).key_to_index):
                if i >= size:
                    break
                results.append(word)
            return results
        return Exception('Unexpected flow')


    def get_dimension(self, source):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')

        return self.get_model(source=source).vectors[0].shape[0]


    def get_vectors(self, source, words=None):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if words is None:
            words = self.get_words(source=source)
        embedding = np.empty((len(words), self.get_dimension(source=source)), dtype=np.float32)
        for i, word in enumerate(words):
            embedding[i] = self.get_vector(source=source, word=word)
        return embedding


    def get_vector(self, source, word):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if source not in self.model:
            raise ValueError('Did not load %s model yet' % source)
        try:
            return self.model[source][word]
        except KeyError as e:
            dims = self.model[source][0].shape
            vect = np.empty(dims)
            vect[:] = np.nan
            return vect


    def get_synonym(self, source, word, topn=5):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if source not in self.model:
            raise ValueError('Did not load %s model yet' % source)
        try:
            return self.model[source].most_similar(positive=word, topn=topn)
        except KeyError as e:
            raise


    def get_distance_between_two_words(self, source, word1, word2):
        if source not in ['glove', 'word2vec', 'fasttext', 'from_scratch']:
            raise ValueError('Possible value of source are glove, word2vec, fasttext or from_scratch')
        if source not in self.model:
            raise ValueError('Did not load %s model yet' % source)
        try:
            return self.model[source].similarity(word1, word2)
        except KeyError as e:
            raise
```
